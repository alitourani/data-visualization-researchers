```python
# Required Libraries
import pandas as pd

# Reading StackOverflow data
dataFrame = pd.read_csv('Dataset.csv')
dataFrame.head()
```

```python
# Processing Developers' Nationalities Data
nationalityData = dataFrame['Country'].value_counts().rename_axis('Country').reset_index(name='Programmers')
# nationalityData.head()

nationalityData = nationalityData.replace('Russian Federation', 'RU')
nationalityData = nationalityData.replace('Hong Kong (S.A.R.)', 'Hong Kong')
nationalityData = nationalityData.replace('Iran, Islamic Republic of...', 'Iran')
nationalityData = nationalityData.replace("Lao People's Democratic Republic", 'Laos')
nationalityData = nationalityData.replace('United States of America', 'United States')
nationalityData = nationalityData.replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')

# Normalizing the output
# It should be like [['Country', 'Programmers'], ['Germany', 200], ['United States', 300]]
nationalityArray = [['Country', 'Programmers']]
for index, row in nationalityData.iterrows():
    nationalityArray.append([row['Country'], row['Programmers']])
# nationalityArray[0:9]

# Save to file
with open('Nationalities.txt', 'w') as myFile:
    myFile.write('[')
    for item in nationalityArray:
        myFile.write("%s, " % item)
    myFile.write(']')
```

```python
# Processing Developers' Nationalities Data
genderData = dataFrame['Gender'].value_counts().rename_axis('Gender').reset_index(name='Count')

# Note: there are 15+ variations, we want to make it 3: Man, Woman, Other.

totalOthers = genderData.query('Count < 4000').sum()
totalData = pd.DataFrame([['Other', totalOthers['Count']]], columns=['Gender', 'Count'], index=[2])
genderData = pd.concat([genderData.query('Count > 4000'), totalData])
genderData
```
